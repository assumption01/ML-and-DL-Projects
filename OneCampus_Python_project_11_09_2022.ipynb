```python

# Enter the total number of students to record
StudentTotalNumber =int(input("Enter the Total Number Student for Recording: "))
newDict = []
allrec = []

for x in range(StudentTotalNumber):
    print("\n"+"\033[94m" +"New Student Entry"+ "\033[0m")
    studentName            = input("Enter the Student Full name: ")
    studentID              = input("Enter " + studentName + "'s ID: ")

    print("\n"+"\033[95m" + studentName + "'s Chemistry Scores" + "\033[0m")
    studentQuizChem        = int(input("Enter Chemistry Quiz score: "))
    studentHomeworkChem    = int(input("Enter Chemistry Homework scor4e: "))
    studentAttendChem      = int(input("Enter Chemistry Attendance Point: "))
    studentExamChem        = int(input("Enter Chemistry Exams score: "))

    print("\n"+"\033[92m" + studentName + "'s Biology Scores" + "\033[0m")
    studentQuizBio         = int(input("Enter Biology Quiz score: " ))
    studentHomeworkBio     = int(input("Enter Biology Homework score: " ))
    studentAttendBio       = int(input("Enter Biology Attendance Point: " ))
    studentExamBio         = int(input("Enter Biology Exams score: " ))

    print("\n"+"\033[94m" + studentName + "'s Physics Scores" + "\033[0m")
    studentQuizPhy         = int(input("Enter Physics Quiz score: " ))
    studentHomeworkPhy     = int(input("Enter Physics Homework score: " ))
    studentAttendPhy       = int(input("Enter Physics Attendance Point: " ))
    studentExamPhy         = int(input("Enter Physics Exams score: " ))

    print("\n"+"\033[91m" + studentName + "'s Mathematics Scores" + "\033[0m")
    studentQuizMath        = int(input("Enter Mathematics Quiz score: "))
    studentHomeworkMath    = int(input("Enter Mathematics Homework score: "))
    studentAttendMath      = int(input("Enter Mathematics Attendance Point: "))
    studentExamMath        = int(input("Enter Mathematics Exams score: "))

    # Generate Student Average for Each Subject
    chemAverage=(((studentQuizChem/190)*30) + ((studentHomeworkChem/190)*15) + ((studentAttendChem/96)*12) + ((studentExamChem/100)*43))
    bioAverage=(((studentQuizBio/190)*30) + ((studentHomeworkBio/190)*15) + ((studentAttendBio/96)*12) + ((studentExamBio/100)*43))
    phyAverage=(((studentQuizPhy/190)*30) + ((studentHomeworkPhy/190)*15) + ((studentAttendPhy/96)*12) + ((studentExamPhy/100)*43))
    mathAverage=(((studentQuizMath/190)*30) + ((studentHomeworkMath/190)*15) + ((studentAttendMath/96)*12) + ((studentExamMath/100)*43))

    #Student total Average Score
    AverageScores = ((bioAverage + chemAverage + phyAverage + mathAverage) /4)
    
    #Grading
    def grading(AverageScore):
      a="A"; b="B"; c="C"; d="D"; e="E"; f="F"
      if AverageScore >= 90:
        return a
      elif AverageScore < 90 and AverageScore>=75:
        return b
      elif AverageScore < 75 and AverageScore>=65:
        return c
      elif AverageScore < 65 and AverageScore>=55:
        return d
      elif AverageScore < 55 and AverageScore>=50:
        return e
      else:
        return f

    #Status
    def statuss(AverageScore):
        p="PASS"; r="RETAKE"; f1="FAIL"
        if AverageScore <= 100 and AverageScore>=55:
          return p
        elif AverageScore < 55 and AverageScore>=50:
          return r
        else:
          return f1
    grade = grading(AverageScores)
    status = statuss(AverageScores)

    #GPA 
    GPA = ((AverageScores/100) * 5)


#Automatically computed values
    AutoRecords = {"Name": studentName,
                 "ID": studentID, 
                 "Score": AverageScores,
                 "GPA": GPA,
                 "Grade": grade,
                 "Status": status}

# Each studentâ€™s performance records 
    studentRecords = {"Name": studentName,
                 "ID": studentID, 
                 "Quiz": {"Chemistry": studentQuizChem, 
                                      "Biology": studentQuizBio, 
                                      "Physics": studentQuizPhy,
                                      "Maths": studentQuizMath},
                 "Homework": {"Chemistry": studentHomeworkChem, 
                                      "Biology": studentHomeworkBio, 
                                      "Physics": studentHomeworkPhy,
                                      "Maths": studentHomeworkMath},
                 "Attendance": {"Chemistry": studentAttendChem, 
                                      "Biology": studentAttendBio, 
                                      "Physics": studentAttendPhy,
                                      "Maths": studentAttendMath},
                 "Exams": {"Chemistry": studentExamChem, 
                                      "Biology": studentExamBio, 
                                      "Physics": studentExamPhy,
                                      "Maths": studentExamMath}}

  #Each subject performance records for all students
    subjectsRecord = {"Name": studentName,
                 "ID": studentID,
                 "Chemistry": {"Quiz": studentQuizChem, 
                                      "Homework": studentHomeworkChem, 
                                      "Attendance": studentAttendChem,
                                      "Exams": studentExamChem,
                                      "AverageScore": chemAverage},
                 "Biology": {"Quiz": studentQuizBio, 
                                      "Homework": studentHomeworkBio, 
                                      "Attendance": studentAttendBio,
                                      "Exams": studentExamBio,
                                      "AverageScore": bioAverage},
                 "Physics": {"Quiz": studentQuizPhy, 
                                      "Homework": studentHomeworkPhy, 
                                      "Attendance": studentAttendPhy,
                                      "Exams": studentExamPhy,
                                      "AverageScore": phyAverage},
                 "Mathematics": {"Quiz": studentQuizMath, 
                                      "Homework": studentHomeworkMath, 
                                      "Attendance": studentAttendMath,
                                      "Exams": studentExamMath,
                                      "AverageScore": mathAverage}}

    newDict.append(subjectsRecord)

#All Students Records
    allRecords = {"Name": studentName,
                 "ID": studentID, 
                 "Chem Quiz": studentQuizChem, "Bio Quiz": studentQuizBio, "Phy Quiz": studentQuizPhy, "Math Quiz": studentQuizMath,
                 "Chem HW": studentHomeworkChem, "Bio HW": studentHomeworkBio, "Phy HW": studentHomeworkPhy, "Math HW": studentHomeworkMath,
                 "Chem Attnd": studentAttendChem, "Bio Attnd": studentAttendBio, "Phy Attnd": studentAttendPhy, "Math Attnd": studentAttendMath,
                 "Chem Exam": studentExamChem, "Bio Exam": studentExamBio, "Phy Exam": studentExamPhy,"Math Exam": studentExamMath,
                 "Chem AVG": chemAverage,
                 "Bio AVG": bioAverage,
                 "Phy AVG": phyAverage,
                 "Math AVG": mathAverage,                
                 "Avg Score": AverageScores,
                 "GPA": GPA,
                 "Grade": grade,
                 "Status": status}
allrec.append(allRecords)


print("\n"+"\033[91m" +"Student Records"+ "\033[0m")
print(studentRecords)

print("\n"+"\033[92m" +"Records of Subjects for all Student"+ "\033[0m")
print(newDict)

print("\n"+"\033[94m" +"Auto generated Grades for Each student"+ "\033[0m")
print(AutoRecords)

print("\n"+"\033[92m" +"All student record"+ "\033[0m")
print(allrec)
```
